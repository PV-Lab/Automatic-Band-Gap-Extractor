## Example usage of the Automatic Band Gap Extractor package

#### The "data" input should be in the format of an (n x m) pandas array with n reflectance data points and (m - 1) measured spectra, where m = 0 is the wavelength

Package written by Alexander (Aleks) E. Siemenn <asiemenn@mit.edu>, 05 Feb. 2023

```python
import os
import pandas as pd
import extractor
```

```python
datapath = os.getcwd() + '\\example-data\\'
data = pd.read_csv(datapath + 'example-data.csv')
display(data) # diaply of sample data input format with labeled headers for each measured sample and wavelength values in the first column
```

```python
extractor.autoextract(data=data, csvname='example-output',
                      savepath=datapath+'example-output\\', verbose=True)
```
